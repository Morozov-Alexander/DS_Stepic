```python
import matplotlib as plt
import pandas as pd
from sklearn import tree
import numpy as np
%matplotlib inline
```

```python
dogs = pd.read_csv('dogs.csv')
cats = pd.read_csv('cats.csv')

cats
```

```python
cs = tree.DecisionTreeClassifier(criterion = 'entropy')
cs.fit(cats[['Шерстист']],cats['Вид'])
```

```python
tree.plot_tree(cs, feature_names=list(cats[['Шерстист']]),
               class_names=['собачка', 'котик'],
               filled=True)
```

```python
entropy = - 5/9 * np.log2(5/9) - 4/9 * np.log2(4/9)
entropy
```

```python
cs = tree.DecisionTreeClassifier(criterion = 'entropy')
cs.fit(cats[['Гавкает']],cats['Вид'])
tree.plot_tree(cs, feature_names=list(cats[['Гавкает']]),
               class_names=['собачка', 'котик'],
               filled=True);
```
